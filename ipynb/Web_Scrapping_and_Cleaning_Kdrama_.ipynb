```python
import requests
from bs4 import BeautifulSoup

Title = []
Year = []
Score = []
Sinopsis = []
URL = []
Cast =[]
Rating = []
Aired = []
Genre = []
Tags = []
for i in range(1,250):
  url = f'https://mydramalist.com/shows/top?page={i}'
  response = requests.get(url)
  response
  soup = BeautifulSoup(response.text, 'html.parser')
  html = soup.find_all('div', {"class" : 'col-xs-9 row-cell content'})
  
  for drama in html:
    title = drama.find('a', href=True).text
    Title.append(title)

    year = drama.find("span", "text-muted").text
    Year.append(year)

    score = drama.find("span", "p-l-xs score").text
    Score.append(score)

    drama_url = 'https://mydramalist.com' + drama.find('a')['href']
    response_drama = requests.get(drama_url)
    detail_drama = BeautifulSoup(response_drama.text, 'html.parser')

    #Find informastion on detail tab
    details = detail_drama.find_all('div', {'class': 'show-synopsis'})
    for detail in details :
      sinopsis = [teks.get_text() for teks in detail.find_all('span')]
      Sinopsis.append(sinopsis)

    #Find Images 
    images = detail_drama.find_all('div', 'col-sm-4 film-cover cover')
    for image in images : 
      link = image.find('img', {'class': 'img-responsive'})['src']
      URL.append(link)
    #Cast
    casts = detail_drama.find_all('div', 'p-a-sm')
    for cast in casts:
      names = [name.get_text() for name in cast.find_all('b')]
      Cast.append(names)
    #Detail 2
    rates = detail_drama.find_all('li', 'list-item p-a-0 content-rating')
    for rate in rates :
      rating = rate.text.strip().split(': ')[1]
      Rating.append(rating)
    #Watch
    aired = detail_drama.find_all('div','box-body wts')
    for air in aired :
      try :
        watch = [tv.get_text() for tv in air.find_all('b')]
      except : 
        watch = ' '
    Aired.append(watch)
    #Genre
    genres = detail_drama.find_all('li', 'list-item p-a-0 show-genres')
    for genre in genres:
      try :
        gen = [g.get_text() for g in genre.find_all('a', href=True)]
      except :
        gen = ' '
    Genre.append(gen)

    #Tags
    tags = detail_drama.find_all('li', 'list-item p-a-0 show-tags')
    for tag in tags :
      try :
        name_tag = [name_tags.get_text() for name_tags in tag.find_all('span')]
      except :
        name_tag = ' '
      Tags.append(name_tag)
```

```python
import pandas as pd
df = pd.DataFrame({"Name": Title,
                  "Year" : Year,
                  "Genre" : Genre,
                  "Main Cast" : Cast,
                  "Sinopsis": Sinopsis,
                  "Score": Score,
                  "Content Rating" : Rating,
                   "Tags" : Tags,
                  "Network" : Aired,
                  "img url" : URL})
df
```

```python
df.to_csv('Dramalist raw.csv')
```

```python
data = pd.read_csv('Dramalist raw.csv')

data.head()
```

```python
data[['Type', 'Year', 'Episode']] = data['Year'].str.split(' - |, ', expand=True)
data.head()
```

```python
data = data[data['Type'] == 'Korean Drama'].reset_index(drop=True)
data = data.drop('Type', axis =1)
data = data.drop('Unnamed: 0', axis=1)
data
```

```python
def clean_simbol(kolum):
  data[kolum]=data[kolum].apply(lambda x: str(x).replace('[','').replace(']','').replace('\'','').replace('()', ' ').replace('~~', ' '))
  return data[kolum]
```

```python
clean_simbol('Genre')
clean_simbol('Main Cast')
clean_simbol('Network')
clean_simbol('Sinopsis')
clean_simbol('Tags')
data.head()
```

```python
import re
def clean_sinopsis(sinopsis):
    sinopsis = re.sub(r'Aired.+',' ',sinopsis )
    sinopsis = re.sub(r'Release dates:.+', '', sinopsis)#Menghapus informasi yang tidak berkaitan
    sinopsis = re.sub(r'\n+', '\n', sinopsis)  # menggabungkan baris yang terlalu banyak \n
    
    # menghilangkan karakter \n dan koreksi penulisan
    sinopsis = sinopsis.strip()
    sinopsis = re.sub('[\u3131-\u3163\uac00-\ud7a3]+', '', sinopsis)#menghapus tulisan hangul
    sinopsis = re.sub(r'\(Source:.*?\)\s*', ' ',sinopsis)
    sinopsis = re.sub(r'\\n', ' ', sinopsis)
    sinopsis = re.sub(r' +', ' ', sinopsis)
    
    return sinopsis
```

```python
data['Sinopsis'] = data['Sinopsis'].apply(clean_sinopsis)
data.head()
```

```python
data.to_csv('kdrama_list.csv')
```
