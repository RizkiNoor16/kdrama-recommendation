```python
import pandas as pd
df = pd.read_csv('kdrama_list.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Main Cast']= df['Main Cast'].apply(lambda x: ' '.join(x.split(', ')[:3]))
df.head()
```

```python
df.head()
```

```python
df['overview']=df['Tags']+' - ' + df['Content Rating'] + ' - '+ df['Genre'] +' - ' + df['Main Cast']
df['overview']
```

```python
data = df.drop(columns=['Sinopsis', 'Content Rating', 'Genre', 'Network', 'Main Cast', 'Unnamed: 0','Year', 'Score', 'Episode', 'Tags'])
data.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
data['overview']= data['overview'].apply(stem)
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfid = TfidfVectorizer(stop_words='english')
```

```python
feature_matrix = tfid.fit_transform(data['overview'])
```

```python
feature_matrix
```

```python
similarity = cosine_similarity(feature_matrix)
print(similarity.shape)
```

```python
def recommend_drama(title, cosine_sim=similarity, df=df):
    # mencari indeks drama berdasarkan judul
    idx = df[df['Name'] == title].index[0]
    
    # mengambil nilai cosine similarity untuk drama yang dipilih
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # mengurutkan drama berdasarkan nilai cosine similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # mengambil 10 drama dengan nilai cosine similarity tertinggi
    sim_scores = sim_scores[1:11]
    drama_indices = [i[0] for i in sim_scores]
    
    # mengembalikan judul drama yang direkomendasikan
    return df[['Name','img url']].iloc[drama_indices]
```

```python
recommend_drama('Goblin')
```
